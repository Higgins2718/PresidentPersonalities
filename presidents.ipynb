```
import pandas as pd
from pandas.io.json import json_normalize
personality_data = json_normalize(data=json_dict['personality'])
personality_data = personality_data.drop('trait_id', 1)
personality_data = personality_data.drop('category', 1)
personality_data = personality_data.drop('children', 1)


personality_data.head()

```

```
df = pd.DataFrame({'Name': "John Adams", 'Openness':  personality_data['percentile'][0],
                   'Openness_significant':  personality_data['significant'][0],

                  'Conscientiousness': personality_data['percentile'][1],
                  'Conscientiousness_significant':  personality_data['significant'][1],

                  'Extraversion': personality_data['percentile'][2],
                  'Extraversion_significant':  personality_data['significant'][2],

                  'Agreeableness': personality_data['percentile'][3],
                  'Agreeableness_significant':  personality_data['significant'][3],

                  'Neuroticism': personality_data['percentile'][4],
                  'Neuroticism_significant':  personality_data['significant'][4],

                   },
                  index=[0])
df.head()
```

```
def append_to_dataframe(json_data, name):
  print(json_data)
  new_df = pd.DataFrame({'Name': name, 'Openness':  json_data['percentile'][0],
                   'Openness_significant':  json_data['significant'][0],

                  'Conscientiousness': json_data['percentile'][1],
                  'Conscientiousness_significant':  json_data['significant'][1],

                  'Extraversion': json_data['percentile'][2],
                  'Extraversion_significant':  json_data['significant'][2],

                  'Agreeableness': json_data['percentile'][3],
                  'Agreeableness_significant':  json_data['significant'][3],

                  'Neuroticism': json_data['percentile'][4],
                  'Neuroticism_significant':  json_data['significant'][4],

                   },
                  index=[0]).append(df, ignore_index=True)
  return new_df
df = append_to_dataframe(personality_data, 'Theodore Roosevelt')
df.head()
```
